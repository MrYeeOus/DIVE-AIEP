```python
import pandas as pd
import json

from datasets import load_dataset
```

```python
with open("mergedJson.jsonl", "r") as jsonl_file:
    lines = jsonl_file.readlines()
    noot = jsonl_file.read()

data_list = []
for line in lines:
    data = json.loads(line)
    data_list.append(data)

dataframe = pd.DataFrame(data_list, columns=["GENERATED_DESCRIPTION", "GENERATED_DATA"])
# dataframe = pd.DataFrame.from_dict(data_list, orient='columns')

dataframe
```

```python
# !pip install nltk
# import nltk
# nltk.download("punkt")

from nltk.tokenize import word_tokenize

dataframe["GENERATED_DESCRIPTION"] = dataframe["GENERATED_DESCRIPTION"].astype(str)
dataframe["GENERATED_DATA"] = dataframe["GENERATED_DATA"].astype(str)
dataframe["TOKENIZED_DESCRIPTION"] = dataframe["GENERATED_DESCRIPTION"].apply(word_tokenize)
dataframe["TOKENIZED_DATA"] = dataframe["GENERATED_DATA"].apply(word_tokenize)
```
